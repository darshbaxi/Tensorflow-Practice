```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img =cv2.imread('/content/picai.pngai.png')
cv2_imshow(img)
```

```python
## Scalinging
img=cv2.resize(img,(300,300),interpolation=cv2.INTER_CUBIC)
cv2_imshow(img)
```

```python
## Translation
import numpy as np
```

```python
def translate(img,x,y):
  transmat=np.float32([[1,0,x],[0,1,y]])
  translated_image = cv2.warpAffine(img, transmat, (img.shape[1], img.shape[0]))
  return translated_image
```

```python
img1=translate(img,50,50)
cv2_imshow(img1)
print('original image')
cv2_imshow(img)
```

```python
## filpling
img_flip_horiz=cv2.flip(img1,0) ##0 DENOTE FOR FLIP ALONG X-AXIS
cv2_imshow(img_flip_horiz)
print('original image')
cv2_imshow(img1)
```

```python
img_flip_verti=cv2.flip(img1,1) ## 1 DENOTE FOR FLIP ALONG Y-AXIS
cv2_imshow(img_flip_verti)
print('original image')
cv2_imshow(img1)
```

```python
img_flip_horiz_vert=cv2.flip(img1,-1) ## -1 DENOTE FOR FLIP ALONG BOTH(X AND Y) AXIS
cv2_imshow(img_flip_horiz_vert)
print('original image')
cv2_imshow(img1)
```

```python
## ROTATION
def rotation(imga,cen,ang,scal):
  rotate_matrix = cv2.getRotationMatrix2D(center=cen, angle=ang, scale=scal)
  rotated_image = cv2.warpAffine(src=imga, M=rotate_matrix, dsize=(imga.shape[1], imga.shape[0]))
  return rotated_image
```

```python
center=(img.shape[0]/2,img.shape[1]/2)
```

```python
cv2_imshow(rotation(img,center,-45,1))
```
